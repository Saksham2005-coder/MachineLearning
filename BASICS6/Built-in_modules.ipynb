# What are Modules?

- **Code libraries** in Python.
- **Contain functions** for use.

*Examples:*
- **Math**: Math ops
- **Random**: Random numbers
- **os**: OS interaction
- **time**: Time tasks

```python
# keyword module
import keyword
print(keyword.kwlist)  # Reserved keywords in Python
```

### Math Module

```python
# include <stdio.h>
import math
```

```python
math.pi
```

```python
math.e
```

```python
math.factorial(5)
```

```python
math.ceil(6.3)
```

```python
math.floor(6.9)
```

### Random Module

```python
import random
```

```python
random.randint(1, 100)
```

```python
a = [1, 2, 3, 4, 5]
random.shuffle(a)
a
```

### Time Module

```python
import time
```

```python
time.time() # Current time (seconds)
```

```python
time.ctime() # Current time (string)
```

```python
print("Hello")
time.sleep(5) # Delay for 5 seconds
print("World")
```

### Datetime Module

```python
import datetime
print(datetime.datetime.now()) # Current date/time
```
